```python
import pydiffarti as pd
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt
```

```python
anymal_urdf_path = Path("/home/sgiles/anymal_ws/src/anymal_c_simple_description/urdf/anymal2.urdf")
plane_urdf_path = Path("/home/sgiles/src/tiny-differentiable-simulator/data/plane_implicit.urdf")
assert anymal_urdf_path.exists()
assert plane_urdf_path.exists()

g = pd.Utils.fraction(-981, 100)
dt = 1/100

world = pd.TinyWorld(gravity_z=g, do_vis=False)
world.dt = pd.Utils.scalar_from_double(dt)

parser = pd.TinyUrdfParser()
convert_tool = pd.UrdfToMultiBody2()

plane_mb = world.create_multi_body()
plane_mb.isFloating = False
plane_urdf_data = parser.load_urdf(str(plane_urdf_path))
convert_tool.convert2(plane_urdf_data, world, plane_mb)

mb = world.create_multi_body()
mb.isFloating = True
urdf_data = parser.load_urdf(str(anymal_urdf_path))
convert_tool.convert2(urdf_data, world, mb)

mb_solver = pd.TinyMultiBodyConstraintSolver()
dispatcher = world.get_collision_dispatcher()


mb_solver.cfm_ = pd.fraction(1, 500)
mb_solver.erp_ = pd.fraction(1, 5)
mb_solver.pgs_iterations_ = 50
mb_solver.keep_all_points_ = False
world.friction = pd.TinyDualDouble(1.8,0)
```
